```python
import pandas as pd
```

```python
def setID(df,num):
    id = []
    for i in range(len(df)):
        id.append(f'ref{num}_{i}')
    df['ID'] = id
    df.reset_index(drop=True, inplace=True)
    return df
```

-----

## Acute Oral Toxicity

```python
training = pd.read_csv('./Acute Oral Toxicity/Predictive Models for Acute Oral Systemic Toxicity/trainingset_171130.tsv', sep='\t')
test = pd.read_csv('./Acute Oral Toxicity/Predictive Models for Acute Oral Systemic Toxicity/validationset.tsv', sep ='\t')
df = pd.concat([training,test])
df = setID(df,1)[['ID','Canonical_QSARr','very_toxic','nontoxic','LD50_mgkg','EPA_category','GHS_category']]
df
```

```python
df.to_csv('./Acute Oral Toxicity/AOT.csv')
```
