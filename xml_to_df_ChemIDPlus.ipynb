```python
import xml.etree.ElementTree as elemTree
import pandas as pd
import numpy as np
```

```python
tree = elemTree.parse('./CurrentChemID.xml')
root = tree.getroot()
```

```python
df_cols = ["ID", "Formula", "Name", "Name Source", "Synonyms", "Synonyms Source", "CAS"]
rows = []

for node in root:
    chem_id = node.attrib.get('id')
    chem_formula = node.attrib.get('displayFormula')
    chem_name = node.attrib.get('displayName')
    try:
        source = node.find('NameList').find('NameOfSubstance').find('SourceList').find('Source').text
    except:
        source = "-"
    try:
        synonyms = node.find('NameList').find('Synonyms').text
    except:
        synonyms = "-"
    try:
        synSource = node.find('NameList').find('Synonyms').find('SourceList').find('Source').text
    except:
        synSource = "-"
    try:
        cas = node.find('NumberList').find('CASRegistryNumber').text
    except:
        cas = "-"

    rows.append({"ID":chem_id, "Formula":chem_formula, "Name":chem_name, "Name Source":source,'Synonyms':synonyms, "Synonyms Source":synSource, "CAS":cas})

out_df = pd.DataFrame(rows, columns=df_cols)
out_df
```
